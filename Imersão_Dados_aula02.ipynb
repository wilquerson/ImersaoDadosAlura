# **Aula 02 | Estatísticas, visualização de dados e distribuições**

```python
import pandas as pd

url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

dados = pd.read_csv(url_dados, compression = 'zip')
dados
```

```python
mapa = {'droga': 'composto'}
dados.rename(columns = mapa, inplace=True)
```

```python
dados.head()
```

```python
compostos = dados['composto'].value_counts()
compostos.head()
```

```python
cod_compostos = dados['composto'].value_counts().index[:5]
```

```python
cod_compostos
```

```python
dados.query('composto in @cod_compostos')
```

###verificar query na documentação do pandas

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set() # recupera o bug na importação do seaborn
plt.figure(figsize=(8, 6))
#sns.countplot(x = 'composto' , data = dados.query('composto in @cod_compostos')).set_title('Top 5 compostos')
ax = sns.countplot(x = 'composto' , order= cod_compostos, data = dados.query('composto in @cod_compostos'))
ax.set_title('Top 5 compostos')
ax.set_ylabel('Composto')
ax.set_ylabel('Quantidade')
plt.show()
```
